```python
#=============================================================================================================================================#
#FIFA Rank File:https://www.kaggle.com/datasets/cashncarry/fifaworldranking?resource=download
#The files used in the code are "fifa_ranking-2022-12-22.csv","ability_18/22.csv" (encoding="utf-8").
#The files used in the code are "2022/2018 World Cup Player.csv" (encoding="utf-8-sig").
#The code is to generate "2018/22 Team Ability.csv" (encoding=''ufg-8-sig'), which contains the ability and rank of the whole world cup teams
#The lack of columns in file "2018/22 Team Ability.csv" will be directly modified from .csv file to replace with mean value
#=============================================================================================================================================#
```

```python
import pandas as pd
pd.set_option('display.max_columns', 500)
```

```python
def ability_search(nation="France",player_list=["Manuel Peter Neuer","Adam Kelsey","Leslie Sackey"],year=2022):
    print("\nNation: ",nation)
    rank=pd.read_csv("fifa_ranking-2022-12-22.csv",encoding="UTF-8")        
    #Search the fifa rank of the team
    if year==2018:
        df=pd.read_csv("ability_18.csv",encoding = "utf-8")
        rank=rank[rank["rank_date"]<"2018-06-14"]
    elif year==2022:
        df=pd.read_csv("ability_22.csv",encoding = "utf-8")
        rank=rank[rank["rank_date"]<"2022-11-20"]

    if rank["country_full"].str.contains(nation).any():
        home_rank=(rank[rank["country_full"]==nation]["rank"].iloc[-1])
        rank_frame=pd.DataFrame({"Rank":[home_rank]})
    else:
        print(nation,": Not found")
    
    sum=pd.DataFrame()
    column_means=pd.DataFrame()
    for name in player_list:
        name=name.replace('-',' ')
        if df["long_name"].str.contains(name).any():
            form=df[df["long_name"]==name].iloc[:,2:]                # if year is 2022 change 1 -> 2
            sum=pd.concat([sum,form],axis=0,ignore_index=True)
        else:
           print(name," :Not found")
    sum=sum.fillna(0)
    sum = sum.apply(pd.to_numeric, errors='coerce')
    column_means = sum[sum != 0].mean().to_frame().transpose()
    column_means=pd.concat([column_means,rank_frame],axis=1)
      
    return column_means
```

```python
player=pd.read_csv("2022 World Cup Player.csv",encoding="utf-8-sig",thousands=',')        #Remember change file
result=pd.DataFrame()
for i in range(0,32,1):
    nation=player["Nation"].iloc[i]
    name=pd.DataFrame({"Nation":[nation]})

    player_list=player.iloc[i,1:].dropna().to_list()
    #display(player_list)
    player_list=[x.lower() for x in player_list]
    #print(type(name))
    data=ability_search(nation,player_list)
    #display(data)
    sum=pd.concat([name,data],axis=1)
    #display(sum)
    result=pd.concat([result,sum],axis=0)
    #display(result)
display(result)
display(result.isnull().sum())
#result.to_csv("2018 Team Ability.csv",encoding="utf-8-sig")       #save the file
```
