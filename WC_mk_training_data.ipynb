```python
import pandas as pd
pd.set_option('display.max_columns', 500)
```

```python
def ability_diff(home,away):
    ability=pd.read_csv("2022 Team Ability.csv",encoding="utf-8-sig")               #Remember change file
    if ability["Nation"].str.contains(home).any():
        home_ability=ability[ability["Nation"]==home].drop(["Nation"],axis=1)
    else:
        print(home," not found")
    if ability["Nation"].str.contains(away).any():
        away_ability=ability[ability["Nation"]==away].drop(["Nation"],axis=1)
    else:
        print(away," not found")
    ability_diff=home_ability.reset_index()-away_ability.reset_index()
    return ability_diff.drop(["index"],axis=1).rename(columns={"Rank":"rank_diff"})
```

```python
schedule=pd.read_csv("2022 World Cup Game.csv",encoding="utf-8-sig")               #Remember changer file
schedule=schedule.drop(["year","Group"],axis=1)
display(schedule)
total=pd.DataFrame()
for i in range(0,64):
    home=schedule["Home Team Name"].iloc[i]
    away=schedule["Away Team Name"].iloc[i]
    ability=ability_diff(home,away)
    total=pd.concat([total,ability],axis=0)
result=pd.concat([schedule,total.reset_index()],axis=1).drop(["index"],axis=1)
display(result)
print(result.isnull().sum())
#result.to_csv("2022 training data.csv",encoding="utf-8-sig")                      #save the file
```
