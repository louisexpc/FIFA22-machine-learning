```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 500)
```

```python
players=pd.read_csv("players_18.csv",encoding="UTF-8")                 #remember to change the file
players=players.drop(columns=['release_clause_eur','club_contract_valid_until','club_joined','nation_jersey_number','nation_position','nation_team_id','club_loaned_from','nation_jersey_number','body_type','real_face','player_tags','player_traits','sofifa_id','player_face_url', 'club_logo_url','club_flag_url', 'nation_logo_url', 'nation_flag_url'])
display(players)                              #gkÊ≤íÊúâpace shooting passing dribbling defending physic Êúâ goalkeeping_speed
print(players.isnull().sum())
```

```python
for i in range(0,17954,1):                             #if year is 2022 please change 17954 to 19238
    name=players['player_url'].iloc[i]
    name=name.split("/")[-2].replace("-"," ")
    players['player_url'].iloc[i]=name
print(players["player_url"])
```

```python
ability_label=['player_url' ,'overall'
 , 'age' , 'height_cm', 'weight_kg'
 , 'skill_moves', 'pace',
 'shooting', 'passing', 'dribbling' ,'defending', 'physic',
 'attacking_crossing', 'attacking_finishing', 'attacking_heading_accuracy',
 'attacking_short_passing', 'attacking_volleys' ,'skill_dribbling',
 'skill_curve', 'skill_fk_accuracy' ,'skill_long_passing',
 'skill_ball_control', 'movement_acceleration', 'movement_sprint_speed',
 'movement_agility', 'movement_reactions', 'movement_balance',
 'power_shot_power' ,'power_jumping', 'power_stamina', 'power_strength',
 'power_long_shots', 'mentality_aggression' ,'mentality_interceptions',
 'mentality_positioning', 'mentality_vision' ,'mentality_penalties',
 'mentality_composure' ,'defending_marking_awareness',
 'defending_standing_tackle' ,'defending_sliding_tackle',
 'goalkeeping_diving', 'goalkeeping_handling' ,'goalkeeping_kicking',
 'goalkeeping_positioning' ,'goalkeeping_reflexes', 'goalkeeping_speed']
ability=players[ability_label].fillna(0)
display(ability.isnull().sum())
ability=ability.rename(columns={"player_url":"long_name"})
display(ability)
#ability.to_csv('ability_"year".csv')                 #save the file
```
