```python
correlation=result_18[label].corr()
plt.figure(figsize=(25,25))
sns.heatmap(correlation,annot=True,vmax=1,cmap="Blues")
result_corr=correlation.index[abs(correlation['Result'])>0.3]
print(result_corr)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
y=result_18["Result"]
x=result_18[label].drop(["Result"],axis=1)

X_train,X_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=2023,shuffle=True)
'''
display(X_train)
display(X_test)
display(y_train)
display(y_test)
'''
RFC=RandomForestClassifier()
RFC.fit(X_train,y_train)
score=RFC.score(X_test,y_test)
print(score)
y_22=result_22["Result"]
x_22=result_22[label].drop(["Result"],axis=1)
score=RFC.score(x_22,y_22)
print(score)
display(RFC.predict(x_22))
display(y_22.to_numpy())
```

```python
from sklearn.ensemble import RandomForestRegressor
y=result_18['Goal_dif']
x=result_18[label].drop(["Goal_dif"],axis=1)

X_train,X_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=2023,shuffle=True)
RFR=RandomForestRegressor()
RFR.fit(X_train,y_train)
score=RFR.score(X_test,y_test)
print(score)
y_22=result_22["Goal_dif"]
x_22=result_22[label].drop(["Goal_dif"],axis=1)
score=RFR.score(x_22,y_22)
print(score)
display(RFR.predict(x_22))
display(y_22.to_numpy())
```
